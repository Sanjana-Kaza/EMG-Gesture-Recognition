```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
from vispy.plot import Fig


samp_freq = 500  # Sample frequency (Hz)
notch_freq = 60.0  # Frequency to be removed from signal (Hz)
quality_factor = 30.0  # Quality factor
b_notch, a_notch = signal.iirnotch(notch_freq, quality_factor, samp_freq)
outputSignal_A = signal.filtfilt(b_notch, a_notch, data_A)
outputSignal_B = signal.filtfilt(b_notch, a_notch, data_B)
outputSignal_X = signal.filtfilt(b_notch, a_notch, data_X)
outputSignal_Y = signal.filtfilt(b_notch, a_notch, data_Y)

```

```python
b, a = signal.butter(6, 150, btype='lowpass', fs=500)

outputSignal_low_A = signal.filtfilt(b, a, outputSignal_A)
outputSignal_low_B = signal.filtfilt(b, a, outputSignal_B)
outputSignal_low_X = signal.filtfilt(b, a, outputSignal_X)
outputSignal_low_Y = signal.filtfilt(b, a, outputSignal_Y)
```

```python
b, a = signal.butter(6, 40, btype='highpass', fs=500)

outputSignal_high_A = signal.filtfilt(b, a, outputSignal_low_A)
outputSignal_high_B = signal.filtfilt(b, a, outputSignal_low_B)
outputSignal_high_X = signal.filtfilt(b, a, outputSignal_low_X)
outputSignal_high_Y = signal.filtfilt(b, a, outputSignal_low_Y)
```
